```python
# Import pandas for data manipulation
import pandas as pd

# Import matplotlib for plotting
import matplotlib.pyplot as plt

# Import plotly for interactive visualizations
import plotly.express as px
import seaborn as sns
```

```python

#show 5 first rows
df1 = pd.read_csv("data/properati-MX-2017-07-01-properties-sell.csv")
df1.head()
```

```python
df1.info()       # column types + missing values
```

```python
df1.shape        # rows and columns  # first 5 rows
```

```python
df1.describe()   # summary statistics
```

```python
df1.isnull().sum()  # missing values per column
```

```python
df1 = df1.drop_duplicates()
```

```python
df1.head()     # rows and columns  # first 5 rows
```

```python
df1.rename(columns={
    "id": "ID",
    "properati_url": "Properati_url",
    "description": "Description",
    "title": "Title",
    "created_on": "Created_on",
    "operation": "Operation",
    "place_name": "Place_name",
    "property_type": "Property_type",
    "country_name": "Country_name",
    "state_name": "State_Name",
    "price_usd_per_m2": "Price_usd_per_m2",
    "price_per_m2": "Price_per_m2",
    "floor": "Floor",
    "rooms": "Rooms",
    "expenses": "Expenses"
}, inplace=True)
```

```python
df1.head()
```

```python
### summary
summary = {
    "avg_price_per_m2": df1["Price_per_m2"].mean(),
    "max_price_per_m2": df1["Price_per_m2"].max(),
    "min_price_per_m2": df1["Price_per_m2"].min()
}

summary
```

```python
### Scatter Plot (Price vs Area)


plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df1,
    x="surface_covered_in_m2",
    y="Price_usd_per_m2",
    color="steelblue"
)

plt.title("Relationship Between House Area and Price")
plt.xlabel("Area (Square Meters)")
plt.ylabel("Price (USD)")
plt.grid(True, alpha=0.3)

plt.show()
```

```python
### Distribution of Price
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df1,
    x="surface_covered_in_m2",
    y="Price_usd_per_m2",
    color="steelblue"
)

plt.title("Relationship Between House Area and Price")
plt.xlabel("Area (Square Meters)")
plt.ylabel("Price (USD)")
plt.grid(True, alpha=0.3)

plt.show()
```

```python
### Distribution of Price
plt.figure(figsize=(8,5))

sns.histplot(
    data=df1,
    x="surface_covered_in_m2",
    y="Price_usd_per_m2",
    color="steelblue"
)

plt.title("Relationship Between House Area and Price")
plt.xlabel("Area (Square Meters)")
plt.ylabel("Price (USD)")
plt.grid(True, alpha=0.3)

plt.show()
```

```python
## 7. Correlation Analysis

### Correlation Matrix

corr = df1.corr(numeric_only=True)
corr

```

```python
### Specific Correlation (Price vs Area)


df1["Price_per_m2"].corr(df1["surface_covered_in_m2"])

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Larger houses tend to have higher prices (positive correlation between area and price)
# Some properties are extreme outliers (very high-priced houses)
# Price distribution is skewed (most houses are in lower to mid-range prices)
# Number of rooms may have weaker correlation than area*
```
